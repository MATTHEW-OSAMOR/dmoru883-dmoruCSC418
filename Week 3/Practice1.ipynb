```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# Load the image
img = cv.imread('Goat.png', 0)  # Load in grayscale
rows, cols = img.shape

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

# Define the transformation matrix for translation
M = np.float32([[1, 0, 100], [0, 1, 50]])  # Move 100 pixels right and 50 pixels down
transl = cv.warpAffine(img, M, (cols, rows))

# Plot the translated image
plt.subplot(1, 2, 2)
plt.title("Translated Image")
plt.imshow(transl)

plt.show()
```
