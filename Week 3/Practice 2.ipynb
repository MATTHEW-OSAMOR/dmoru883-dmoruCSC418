```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

# Load the image
img = cv.imread('Goat.png', 0)  # Load image
rows, cols = img.shape  # Get image dimensions

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

# Define the reflection transformation matrix
M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
reflected_img = cv.warpPerspective(img, M, (int(cols), int(rows)))

# Plot the reflected image
plt.subplot(1, 2, 2)
plt.title("Reflected Image")
plt.imshow(reflected_img)

# Save the reflected image
cv.imwrite('reflection_out.jpg', reflected_img)

cv.waitKey(0)
cv.destroyAllWindows()
```
